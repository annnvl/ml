```python
data = pd.read_csv('task2_lemmas_train')
test = pd.read_csv('task2_lemmas_test')
```

```python
olddata = data.copy()
data = data[['X','y']]
```

```python
data['PoS'] = data['y'].map(lambda x: x[-1:])
data['y'] = data['y'].map(lambda x: x[:-2])
data.head()
```

```python
exp_data = data.copy()
for x in exp_data.iterrows():
    i = 0
    while(x[1]['X'][i] == x[1]['y'][i]):
        i+=1
        if(i == len(x[1]['y']) or i == len(x[1]['X'])):
            x[1]['X'] = x[1]['y'] = ''
            break
    
    if(i < len(x[1]['y'])):
        x[1]['X'] = x[1]['X'][i:]
        x[1]['y'] = x[1]['y'][i:]
```
